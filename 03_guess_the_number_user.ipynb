```python
import random


def guess(x):
  number  = random.randint(1,100)
  guess = 0
  while guess != number:
    guess = int(input(f"Guess a number between 1 and {x}: "))
    if guess < number:
      print("Sorry, guess again. Too low.")
    elif guess > number:
      print("Sorry, guess again. Too high.")
  print(f"Congrats. You have guessed the number {number} correctly!!")

guess(100)

```
