```python
import streamlit as st
import pandas as pd

st.title('Streamlit Practice')

height = st.slider('Select your height (in cm):', 50, 250, 150)
weight = st.slider('Select your weight (in kg):', 20, 250, 70)

bmi = weight / (height / 100) ** 2

st.write(f"Your BMI is {bmi:.2f}")
st.write("BMI Category: ")
if bmi < 18.5:
    st.write("Underweight")
elif bmi < 25:
    st.write("Normal")
elif bmi < 30:
    st.write("Overweight")
else:
    st.write("Obese")


```
